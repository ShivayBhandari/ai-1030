 #matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
x=np.array([0,6])
y=np.array([0,100])
plt.plot(x,y)
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='*')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='*',linestyle='dotted')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='*',linestyle='dashed')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',linestyle='dashdot')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
xpoints=np.array([4,6,10,12])
plt.plot(ypoints,xpoints)
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',linestyle='dashdot',color="r")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',linestyle='dashdot',color="hotpink")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',color="r",linewidth="15.0")
plt.show()
```

```python
x=np.array([4,5,7,9,12])
y=np.array([1,3,5,7,9])
plt.plot(x)
plt.plot(y)
plt.show()
```
